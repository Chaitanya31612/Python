Step 4 — Tagging Sentences

```
nltk.download('averaged_perceptron_tagger') #download tagger
from nltk import word_tokenize, pos_tag, pos_tag_sents #import 

tweets_tagged = pos_tag_sents(tweets_tokens) #use
print(tweets_tokens)
```

tweet is represented as a list and for each token we have information about its POS tag. Each token/tag pair is saved as a tuple.

In NLTK, the abbreviation for adjective is JJ.

The NLTK tagger marks singular nouns (NN) with different tags than plural nouns (NNS). To simplify, we will only count singular nouns by keeping track of the NN tag.

In the next step we will count how many times JJ and NN appear throughout our corpus

Step 5 — Counting POS Tags

```
#The first loop will iterate through each tweet in the list. The second loop will iterate through each token/tag pair in each tweet. 
# For each pair, we will look up the tag using the appropriate tuple index.
JJ_count = 0
NN_count = 0

for tweet in tweets_tagged:
    for pair in tweet:
        tag = pair[1]
        if tag == 'JJ':
            JJ_count += 1
        elif tag == 'NN':
            NN_count += 1
print('Total number of adjectives = ', JJ_count)
print('Total number of nouns = ', NN_count)
```
